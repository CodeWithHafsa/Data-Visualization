### Read csv file using pandas

```python
import pandas as pd
who = pd.read_csv("WHO.csv", parse_dates=["Date_reported"])  #parse date to convert date into date time
who.head(5)
```

---

#### Filter to extract italy data from overall data

```python
italy = who[who["Country"] == "Italy"]  #filter to extract italy data from overall data
italy
```

```python
who.head(5)
```

```python
who.info()
```

---

### Line chart between Date Reported and Cumulative Class for Italy data
Here:
* x is independent variable
* y is dependent variable

```python
# here matplotlib is library and pyplot is its class and plt stand for plot
import matplotlib.pyplot as plt 
plt.title("Italy: Covid Cumulative Cases")
plt.xlabel("Date")
plt.ylabel("Cumulative Cases")
plt.plot(italy.Date_reported, italy.Cumulative_cases)
plt.show()
```

### Graphs / Regression:
* Growth graphs - Exponential growth, Linear growth, Logarithm growth, Logistic growth
* Decline graphs - Exponential decline, Linear decline, Logarithm decline, Logistic decline
* Irregular graph

Key Point: Logistic growth graph starts with - exponential growth and eventually ends at logistic growth.

#### Creating function to extract particular country data from overall data and to plot its graph.

```python
def line_chart(Country):
    c_df = who[who["Country"] == Country]
    plt.title(f"{Country} : Covid Cummulative Cases")    #title of the graph
    plt.xlabel("Date")                                    #x-axis label of the graph
    plt.ylabel("Cummulative Cases")                       #y-axis label of the graph
    plt.ticklabel_format(axis = "Y", style = "plain")     # to remove exponent from y-axis 
    plt.plot(c_df.Date_reported, c_df.Cumulative_cases)
    plt.show()
```

* Graph between Data and Cumulative Cases for India data \
  Result=> Exponential Growth Graph:

```python
line_chart("India")            # graph of India using function 
```

* Graph between Data and Cumulative Cases for India data \
  Result=> Linear Growth Graph:

```python
line_chart("Poland")            # graph of Poland using function 
```

* Graph between Data and Cumulative Cases for India data \
  Result=> Logistic Growth Graph:

```python
line_chart("Italy")            # graph of Italy using function 
```
