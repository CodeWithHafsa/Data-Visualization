## Seabond

```python
import seaborn as sns
# dir(sns)
```

## For Multiple Plotting in one graph

#### Graph between Gr Living Area and Sales Price

```python
# sns.set_theme()     #theme can also set using set-theme method
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice')
plt.show()
```

---

```python
pro["Overall Cond"].value_counts()    #value count of Overall condition
# Note: Result shows that it is categorical data
```

### Graph Plot between Gr Living Area, Sales Price and Overall Condition

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice', 
            hue = "Overall Cond")
plt.show()
```

#### To change the color palette

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice', 
            hue = "Overall Cond", palette = "RdYlGn")
plt.show()
```

---

```python
pro["Rooms"].unique()
# Note: There are two types of rooms according to data
```

### Graph Plot between Gr Living Area, Sales Price and Rooms

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice', 
            hue = "Rooms")
plt.show()
```

---

### Graph Plot between Gr Living Area, Sales Price Overall Condition and Garage Area

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlGn",
           size = 'Garage Area')
plt.show()
```

#### To change the size of circles of Garage Area - Use Sizes

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlGn",
           size = 'Garage Area', sizes = (200,30))

plt.show()
```

---

### Graph Plot between Gr Living Area, Sales Price Overall Condition, Garage Area and Rooms

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlGn",
           size = 'Garage Area', sizes = (1,300),
           style = "Rooms")

plt.show()
```

#### To change markers of  room column

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlGn",
           size = 'Garage Area', sizes = (1,300),
           style = "Rooms", markers = ["^","v"])

plt.show()
```

### Graph Plot between Gr Living Area, Sales Price Overall Condition, Garage Area and Rooms for each year individually

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlGn",
           size = 'Garage Area', sizes = (1,300),
           style = "Rooms", col = 'Year')

plt.show()
```

```python
pro['Land Slope'].unique()

# Note: There are three types of land slope
```

### Graph Plot between Gr Living Area, Sales Price Overall Condition, Garage Area and Rooms for each year and Each Land Slope individually

```python
sns.relplot(data = pro, x = 'Gr Liv Area', y = 'SalePrice',
           hue = "Overall Cond", palette = "RdYlGn",
           size = 'Garage Area', sizes = (1,300),
           style = "Rooms", row = 'Land Slope', col = 'Year',)

plt.show()
```
