```python
fig, ax = plt.subplots(figsize = (4.5,6))

ax.barh(death.Country_Other, death.Total_Deaths, height = 0.45, color = "#b00b1e") # Object oriented interface 
# (height of bar is small and color change to red)
ax.tick_params(bottom = False, left = False) # to remove ticks
for sp_value in ax.spines.values():
    sp_value.set_visible(False)
    
ax.set_xticks([0,150000,300000]) # Set  only 3 values for xaxis
ax.xaxis.tick_top()
ax.set_yticklabels([])  # remove yaxis labels because we want to add our own labels
ax.xaxis.tick_top() 

ax.tick_params(top = False)  # Set the xaxis values on top of graph 
ax.tick_params(axis = "x", colors ="grey")   # Set the axis values color to grey

# plt.title("The Death Toll Worlwide is 1.5+") # Add title on top of the graph
ax.text(x = -80000, y = 23.5, s ="The Death Toll Worlwide is 1.5+", size = 17, weight = "bold")  
# set x and yaxis (means position), size and weight of the graph

ax.text(x = -80000, y = 22.5, s ="Top 20 countries by death toll (December 2020)", size = 10) 
# add subtitle and set its position and size on top of graph
plt.show()
```

Removing spines and ticks on graph\
Set height(means bar size) and color of the bar\
Also small the size of bars, Plus set 3 values for x tick, set this on top and set its color grey\
Remove yaxis labels\
Add Title of the graph (set its position, size and weight)\
Add subtitle of the graph (set its position and size)\
Adding own list to the graph using enumerate

```python
fig, ax = plt.subplots(figsize = (4.5,6))

ax.barh(death.Country_Other, death.Total_Deaths, height = 0.45, color = "#b00b1e") # Object oriented interface 
# (height of bar is small and color change to red)
ax.tick_params(bottom = False, left = False) # to remove ticks
for sp_value in ax.spines.values():
    sp_value.set_visible(False)
    
ax.set_xticks([0,150000,300000]) # Set  only 3 values for xaxis
ax.xaxis.tick_top()
ax.set_yticklabels([])  # remove yaxis labels because we want to add our own labels
ax.xaxis.tick_top() 

ax.tick_params(top = False)  # Set the xaxis values on top of graph 
ax.tick_params(axis = "x", colors ="grey")   # Set the axis values color to grey

# plt.title("The Death Toll Worlwide is 1.5+") # Add title on top of the graph
ax.text(x = -80000, y = 23.5, s ="The Death Toll Worlwide is 1.5+", size = 17, weight = "bold")  
# set x and yaxis (means position), size and weight of the graph

ax.text(x = -80000, y = 22.5, s ="Top 20 countries by death toll (December 2020)", size = 10) 
# add subtitle and set its position and size on top of graph

for i,c in enumerate(countries):
    ax.text(x = -80000, y = i-0.15, s = c) # adding own list in yaxis labels

plt.show()
```

#### Comparision between old graph and updated graph

```python
fig, (ax1, ax2) = plt.subplots(nrows =1, ncols = 2, figsize = (11,6))

ax1.barh(death.Country_Other, death.Total_Deaths) # Object oriented interface
ax1.text(x= -4000, y=23.5,s = "Old graph", size =17, weight = "bold")


ax2.barh(death.Country_Other, death.Total_Deaths, height = 0.45, color = "#b00b1e") # Object oriented interface
ax2.tick_params(bottom = False, left = False) # to remove ticks
for sp_value in ax2.spines.values():
    sp_value.set_visible(False)
    
ax2.set_xticks([0,150000,300000])
ax2.set_xticklabels(["0","150,000","300,000",])
ax2.set_yticklabels([])
ax2.xaxis.tick_top()

ax2.tick_params(top = False)
ax2.tick_params(axis = "x", colors ="grey")
ax2.text(x = -80000, y = 23.5, s ="The Death Toll Worlwide is 150K+", size = 17, weight = "bold")  # plt.title("The Death Toll Worlwide is 1.5+")
ax2.text(x = -80000, y = 22.5, s ="Top 20 countries by death toll (December 2020)", size = 10)

for i,c in enumerate(countries):
    ax2.text(x = -80000, y = i-0.15, s = c)
plt.subplots_adjust(wspace = 0.5)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (4.5,6))

ax.barh(death.Country_Other, death.Total_Deaths, height = 0.45, color = "#b00b1e") # Object oriented interface
ax.tick_params(bottom = False, left = False) # to remove ticks
for sp_value in ax.spines.values():
    sp_value.set_visible(False)
    
ax.set_xticks([0,150000,300000])
ax.set_xticklabels(["0","150,000","300,000",])
ax.set_yticklabels([])
ax.xaxis.tick_top()

ax.tick_params(top = False)
ax.tick_params(axis = "x", colors ="grey")
ax.text(x = -80000, y = 23.5, s ="The Death Toll Worlwide is 150K+", size = 17, weight = "bold")  # plt.title("The Death Toll Worlwide is 1.5+")
ax.text(x = -80000, y = 22.5, s ="Top 20 countries by death toll (December 2020)", size = 10)

for i,c in enumerate(countries):
    ax.text(x = -80000, y = i-0.15, s = c)
ax.axvline(150000, ymin = 0.045, ymax = 1, alpha = 0.5, c = "grey") # adding max or center line in graph
plt.show()
```
