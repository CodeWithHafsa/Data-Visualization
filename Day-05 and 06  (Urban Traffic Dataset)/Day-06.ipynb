##### Zip and Unzip Concept:
To zip file => Use single variable \
To unzip file => Use one or more variable

* E.g. a = 2 , 3 #ans = (2,3) #tuple
* E.g. a,b = 2, 3 #ans => a=2, b=3
* E.g. a,b = 2, 3, 4 #ans => error (because there are three numbers while only two variables)

```python
for i in zip(["Waqas", "Feroz","Hiba"], [2,3,4]):   #zip file
    print(i)
```

```python
for i,j in zip(["Waqas", "Feroz","Hiba"], [2,3,4]):   #to unzip file
    print(i)
```

### Segregate data Manaually

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

traffic_per_day = {}


for day, i in zip(days, range(0,len(df), 27)):
    daily_data = df.iloc[i:i+27]
    
    traffic_per_day[day] = daily_data
```

```python
traffic_per_day["Tue"]   #extract Tuesday Data  #Hour Code start from index 1
```

```python
traffic_per_day["Thu"]   #extract Thursday Data  #Hour Code start from index 1
```

## Individual plot between Hour and Slowness using Pandas:

```python
for day, df in traffic_per_day.items():
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"])
    plt.title(day)
    plt.ylim([0,25]) #we checked that x-axis is same for all the axis so we only set limit for yaxis
    plt.show()
```

### Line chart between Hour and Slowness Using pandas

```python
for day, df in traffic_per_day.items():
    df.plot.line("Hour (Coded)","Slowness in traffic (%)")
    plt.title(day)
    plt.ylim([0,25]) #we checked that x-axis is same for all the axis so we only set limit for yaxis
    plt.show()
```

### Comparison of all days graph in one chart

```python
for day, df in traffic_per_day.items():
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"], label = day)
    
plt.ylim([0,25])
plt.legend()
plt.show()
```
