## Subplot:
It is also known as 'grid chart' or 'small multiples'.\
With the subplot() function you can draw multiple plots in one figure.

#### Syntax:
The subplot() function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot.

plt.subplot(nrows , ncols, index)

```python
# CONCEPT:

# plt.subplot(nrows , ncols, index)
plt.subplot(3 ,2, 2)

plt.subplot(3 ,2, 3)

plt.subplot(3 ,2, 6)

plt.show()
```

### Increase plot size

```python
plt.figure(figsize = (10,12))

# plt.subplot(nrows , ncols, index)
plt.subplot(3 ,2, 2)

plt.subplot(3 ,2, 3)

plt.subplot(3 ,2, 6)

plt.show()
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

traffic_per_day = {}


for day, i in zip(days, range(0,len(df), 27)):
    daily_data = df.iloc[i:i+27]
    
    traffic_per_day[day] = daily_data
```

```python
# Without Comprision
plt.figure(figsize = (10,12))

for i, df in zip(range(0,5), traffic_per_day.values()):
    plt.subplot(3,2, i+1)
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"])
    plt.title(days[i])
    plt.ylim([0,25])
plt.show()
```

```python
# With Comparision
plt.figure(figsize = (10,12))

for i, df in zip(range(0,5), traffic_per_day.values()):
    plt.subplot(3,2, i+1)
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"])
    plt.title(days[i])
    plt.ylim([0,25])
    
plt.subplot(3,2,6)
for day, df in traffic_per_day.items():
    plt.plot(df["Hour (Coded)"], df["Slowness in traffic (%)"], label = day)

plt.ylim([0,25])
plt.title("Comparision")
plt.legend() 
plt.savefig("traffic_variation.png")
plt.show()
```
