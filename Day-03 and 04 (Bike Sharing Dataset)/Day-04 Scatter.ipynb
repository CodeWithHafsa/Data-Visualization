Importing Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

Read File

```python
bike = pd.read_csv("bikeshare/day.csv")
bike.head()
```

---

For categorical data we don't use scatter. If scatter graph is plotted for categorical data, then it prints in a same line as
shown in figure below.

Note: Here scatter graph also shows that registered users are more as compared to casual users in working days.

#### Scatter graph between working day and Casual users - Categorical data

```python
plt.scatter(bike.workingday, bike.casual)
plt.show()

#Note: Registered users line is lesser than working days => Means negative correlation
```

#### Scatter graph between working day and Registered users - Categorical data

```python
plt.scatter(bike.workingday, bike.registered) 
plt.show()

#Note: Registered users line is greater than working days => Means positive correlation
```
