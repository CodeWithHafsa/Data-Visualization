## Seaborn
Seaborn is an amazing visualization library for statistical graphics plotting in Python. It provides beautiful default styles and color palettes to make statistical plots more attractive. It is built on top matplotlib library and is also closely integrated with the data structures from pandas.

Import and Read File

```python
import pandas as pd   #import pandas
import matplotlib.pyplot as plt 
bike = pd.read_csv("bikeshare/day.csv")  #read file
```

Import Seabond

```python
import seaborn as sns
```

```python
# dir(sns)   #directory of seabond
```

```python
sns.countplot(data = bike, x = "yr")    #total sum plot for both years
plt.show()
```

Mapping 1 with Clear. 2 with Cloudy, 3 with Rain in the weather situation data. Skip 4 because it is not in data.

```python
bike["weathersit_cat"] = bike.weathersit.map({1:"Clear", 2:"Cloudy", 3:"Rain"})
bike["weathersit_cat"]
```

```python
# bike.weathersit.astype(str)    #to convert back data into 1,2,3 and 4 form
```

#### Waether Situation Graph for both years individually 

```python
sns.countplot(data = bike,x = "yr", hue = "weathersit_cat")
plt.show()
```

```python
# bike.cnt.unique()        #tells unique value of cnt column         #numpy array
bike.cnt.unique().size     #size of cnt column unique values         #method 1

# len(bike.cnt.unique())     #size of cnt column unique values         #method 2
```

Note: For the above column we can see that uniques values are very large i.e. 696, so we have to make it's frequency table. 
And Histogram is plotted for frequency table data.

```python
bike.cnt.value_counts(bins= 10).sort_index()   
#used to make frequency table 
# bins = 10 (standard)
# sort index - will sort the data
```

### Histogram using Pandas for cnt data

```python
cnt = bike.cnt.value_counts(bins= 10).sort_index() 
plt.hist(bike["cnt"])
plt.show()    #graph shows "Symmetric Normal Distribution"
```

#### Bar Chart Vs Histogram

|Histogram|BarChart|
|------------------------|------------------------|
|Histograms visualize quantitative data or numerical data (discrete or continuous data).|Bar charts display categorical variables. |
|Equal space between every two consecutive bars.|No space between two consecutive bars. They should be attached to each other. |
|Data can be arranged in any order.|Data is arranged in the order of range.|
|The x-axis can represent anything.	|The x-axis should represent only continuous data that is in terms of numbers.|
