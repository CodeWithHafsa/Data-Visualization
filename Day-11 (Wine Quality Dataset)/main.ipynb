```python
postive_white = white_corr > 0    #To get positive values from white wine data
postive_red = red_corr > 0        #To get positive values from red wine data
```

```python
postive_white      #Give answer in bolean
# (True - for positive value, False - for negative value)
```

Map Positive value(true) with one hexcolor and Negative value(False) with another hexcolor - for both white and red wine data

```python
# Map Positive value(true) with one hexcolor and Negative value(False) with another hexcolor
color_white = postive_white.map({True:"#3cb371", False:"#ff0000"})
color_red = postive_red.map({True:"#3cb371", False:"#ff0000"})
```
